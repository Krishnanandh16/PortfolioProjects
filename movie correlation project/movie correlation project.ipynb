```python
#Importing libraries
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Loading the dataset 
df=pd.read_csv("C://Users//krish//OneDrive//Desktop//portfolio project//movies.csv")
#displaying the first 5 rows of dataset
df.head()
```

```python
# checking for the missing values
df.isnull().sum()
```

```python
# replacing the nullvalues
df_1=df.fillna(np.mean(df))
df_1=df_1.dropna()
df_1.isnull().sum()
```

```python
# checking the datatypes of dataframe
df_1.dtypes
```

```python
# changing data types 
df_1['budget']=df_1['budget'].astype('int64')
df_1['gross']=df_1['gross'].astype('int64')
```

```python
df_1['released']=df_1['released'].astype(str)
```

```python
df_1['year_of_release']=df_1['released'].str.extract(pat='([0-9]{4})')
df_1.head()
```

```python
# sorting the data
df_1=df_1.sort_values(by=['gross'],inplace=False,ascending=False)
```

```python
# drop duplicates
df_1.drop_duplicates()
```

```python
#Scatter plot budget vs gross
plt.figure(figsize=(8,6))
plt.scatter(x=df_1['budget'],y=df_1['gross'])
plt.title("Budget vs gross earnings")
plt.xlabel("Gross")
plt.ylabel("budget for flims")
plt.show()
```

```python
# regression plot of budget vs gross
plt.figure(figsize=(8,6))
sns.regplot(x='budget',y='gross',data=df_1,scatter_kws={"color":"red"},line_kws={"color":"black"})
```
